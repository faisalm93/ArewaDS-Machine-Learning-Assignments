In this assignment, I uses the Iris dataset which is a widely used dataset in machine learning. It contains measurements of iris flowers' features along with their corresponding class labels. 
I examined the dataset by creating five scatterplots to visualize the relationships between different pairs of features. The scatterplots compared Sepal Length vs Sepal Width, Petal Length vs Petal Width, Sepal Length vs Petal Length, Sepal Width vs Petal Width, and Sepal Length vs Petal Width.

From the scatterplots, I made several observations and interpretations. In the Sepal Length vs Sepal Width plot, there is some overlap between the Versicolor and Virginica classes, while Setosa had relatively shorter and wider sepals. The Petal Length vs Petal Width scatterplot showed a clear separation between Setosa and the other two classes, with Setosa having shorter and narrower petals. The Sepal Length vs Petal Length plot exhibited a distinct separation between Setosa and the other classes, with longer sepal and petal lengths associated with Versicolor and Virginica.

The Sepal Width vs Petal Width scatterplot revealed that Setosa generally had narrower sepal and petal widths compared to Versicolor and Virginica, with some overlap between the latter two classes. 

Lastly, the Sepal Length vs Petal Width plot showed a partial separation between Setosa and the other classes based on these features. 
Through these scatterplots, I gained insights into the distinct characteristics and variations within the iris species.




```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame from the dataset
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Sepal Length vs Sepal Width:
scatter = plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['target'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset: Sepal Length vs Sepal Width')
```

The scatterplot above shows that there is some overlap between the classes, especially between the Versicolor and Virginica classes. However, we can still notice that Setosa tends to have relatively shorter sepals and wider sepals compared to the other two classes.

```python
# Petal Length vs Petal Width

plt.scatter(df['petal length (cm)'], df['petal width (cm)'], c=df['target'])
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Dataset: Petal Length vs Petal Width')
plt.show()
```

This scatterplot displays the relationship between petal length and petal width. You can observe a clear separation between the Setosa class and the other two classes. Versicolor and Virginica have some overlap.

```python
# Sepal Length vs Petal Length
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'], c=df['target'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Iris Dataset: Sepal Length vs Petal Length')
plt.show()
```

The above scatterplot compares the sepal length with the petal length. You can observe that there is a clear separation between Setosa and the other two classes. Setosa has shorter sepal length and petal length compared to Versicolor and Virginica. Versicolor and Virginica overlap to some extent, but we can still notice a trend where longer sepal length is associated with longer petal length.

```python
# Sepal Width vs Petal Width
plt.scatter(df['sepal width (cm)'], df['petal width (cm)'], c=df['target'])
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Dataset: Sepal Width vs Petal Width')
plt.show()
```

This scatterplot compares the sepal width with the petal width. You can observe that Setosa generally has a narrower sepal width and narrower petal width compared to Versicolor and Virginica. Versicolor and Virginica have a wider range of sepal width and petal width, and there is some overlap between these two classes.

```python
# Sepal Length vs Petal Width
plt.scatter(df['sepal length (cm)'], df['petal width (cm)'], c=df['target'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Dataset: Sepal Length vs Petal Width')
plt.show()
```

This scatterplot examines the relationship between sepal length and petal width. You can observe that there is some separation between Setosa and the other two classes based on these features. Setosa generally has shorter sepal length and narrower petal width compared to Versicolor and Virginica.
