```python

# Import the pyplot module from the matplotlib library
import matplotlib.pyplot as plt

# Define the x-axis data as a list of years
years = [1970, 1980, 1990, 2000, 2010, 2020]

# Define the y-axis data as a list of population values
populations = [100000, 500000, 1000000, 2000000, 3000000, 4000000]

# Create a bar chart using the bar method of pyplot
plt.bar(years, populations)

# Add a label to the x-axis of the chart
plt.xlabel('Year')

# Add a label to the y-axis of the chart
plt.ylabel('Population')

# Add a title to the chart
plt.title('Six Decades Population Trend of Snow Geese')

#The bar chart shows an increase in the population of snow geese over time, with the highest population value of 4,000,000 in the year 2020.

# Display the chart on the screen
plt.show()
```

```python
# Import the required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from a CSV file using pandas
data = pd.read_csv('snow-geese-migration.csv')

# Create a scatter plot using seaborn's scatterplot function
sns.scatterplot(data=data, x='Temperature', y='Migration Date')

# Add a label to the x-axis of the chart
plt.xlabel('Temperature')

# Add a label to the y-axis of the chart
plt.ylabel('Migration Date')

# Add a title to the chart
plt.title('Snow Goose Migration and Temperature')

# Create a lollipop chart using seaborn's lineplot function
sns.lineplot(data=data, x='Temperature', y='Migration Date', sort=True, marker='o')

# Show the chart
plt.show()
#we can observe that as temperature increases, the migration date tends to occur earlier, and as temperature decreases, the migration date tends to occur later.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('snow-geese-vegetation.csv')

# Create a line chart showing the change in Snow Goose diet over time
plt.plot(data['Year'], data['Vegetation'], label='Vegetation')
plt.plot(data['Year'], data['Crops'], label='Crops')

# Add labels and a title to the chart
plt.xlabel('Year')
plt.ylabel('Percentage of diet')
plt.title('Change in Snow Goose diet over time')

# Add a legend to the chart
plt.legend()

# Display the chart
plt.show()
#The chart provides a clear visualization of the change in Snow Goose diet over time where it shows  a decrease in the percentage of diet from vegetation over time, while there has been an increase in the percentage of diet from crops. 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('snow-geese-vegetation.csv')

# Create a line chart showing the change in Snow Goose diet over time
plt.plot(data['Year'], data['Vegetation'], label='Vegetation', color='#75b79e', marker='o', linewidth=2, markersize=8)
plt.plot(data['Year'], data['Crops'], label='Crops', color='#f9ca24', marker='s', linewidth=2, markersize=8)

# Add labels and a title to the chart
plt.xlabel('Year', fontsize=14, labelpad=12, fontweight='bold')
plt.ylabel('Percentage of diet', fontsize=14, labelpad=12, fontweight='bold')
plt.title('Change in Snow Goose diet over time', fontsize=18, pad=20, fontweight='bold')

# Add a legend to the chart
plt.legend(loc='best', fontsize=12, framealpha=0.8)

# Add gridlines to the chart
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Set the size of the figure
plt.figure(figsize=(10, 9))

# Set the x-axis range
plt.xlim([data['Year'].min(), data['Year'].max()])

# Set the y-axis range
plt.ylim([0, 100])

# Add annotations to the data points
for i in range(len(data)):
    plt.annotate(str(data.iloc[i]['Vegetation']) + '%', xy=(data.iloc[i]['Year'], data.iloc[i]['Vegetation'] + 2), ha='center', fontsize=10, fontweight='bold', color='#75b79e')
    plt.annotate(str(data.iloc[i]['Crops']) + '%', xy=(data.iloc[i]['Year'], data.iloc[i]['Crops'] + 2), ha='center', fontsize=10, fontweight='bold', color='#f9ca24')

# Add a background color to the plot
plt.gca().set_facecolor('#f5f5f5')

# Display the chart
plt.show()
```
